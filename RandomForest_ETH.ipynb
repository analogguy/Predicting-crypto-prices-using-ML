```python
import quandl
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

#We use Quandl code to link to Quandl API to load historicald data
#https://www.quandl.com/data/BITFINEX/ETHUSD-ETH-USD-Exchange-Rate
df = quandl.get('BITFINEX/ETHUSD', returns="pandas")
df.head()
```

```python
from ta import add_all_ta_features
from ta.utils import dropna

pd.set_option('display.max_columns', None)

df = dropna(df)

#As all below mentioned features are mandatory in add_all_ta_features we consider low as Open here, as Open price data is not with us from Kraken
df = add_all_ta_features(
    df, open="Low", high="High", low="Low", close="Last", volume="Volume")

df = df.fillna(0)
df.tail()
```

```python
interest = ["Last","Volume", "trend_ema_fast", "momentum_rsi", "others_dlr"]
df2 = df[interest]
df2.describe()




features = ["Volume",  "trend_ema_fast", "momentum_rsi", "others_dlr"]
X = df2[features]
y = df2['Last']
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(random_state=1)
```

```python
rf_model.fit(train_X, train_y)
```

```python
rf_pred = rf_model.predict(val_X)
```

```python
print(rf_model.predict(X.tail()))
```

```python
df2['Last'].tail()
```

```python
from sklearn.metrics import mean_absolute_error
rf_val_mae = mean_absolute_error(val_y,rf_pred)
rf_val_mae
```

#WOW 

```python
#MAE is telling us that on average, our bitcoin price prediction is off by $13.218 from the actual prices.
```

```python
from sklearn.metrics import r2_score
r2_score(val_y, rf_pred)
```

```python
rmse = np.sqrt(((rf_pred - val_y) ** 2).mean())
print('Mean Squared Error: {}'.format(round(rmse, 2)))
```
